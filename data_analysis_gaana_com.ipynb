```python
duration_lst = []
for duration in df['duration']:
  m = int(duration.split(':')[0])
  s = int(duration.split(':')[1])
  sec = m*60 + s

  duration_lst.append(sec)

df['duration'] = duration_lst

df.head()
```

###Average duration of song in each language

```python
for i in lang:
  print(i,int(df[df['language']==i]['duration'].mean()))
```

### No. of singers

```python
all_singers=[]
for singers in df['singer']:
  all_singers+=singers.split('|')
set(all_singers)
```

```python
unique_singers = list(set(all_singers))
```

```python
data = []
for i in unique_singers:
  c=0
  for singers_ in df['singer']:
    if(i in singers_):
      c+=1
  data.append([i,c])
```

```python
from tqdm import tqdm
```

```python
songs = pd.DataFrame(data,columns =['singer','songs'])
```
